# Matplot poate lua si range-uri (ArrayLike)

```python
range_1 = range(4, 8)     # 4,5,6,7,8
range_2 = range(6, 10)    # 6,7,8,9,10

plt.scatter(range_1, range_2);
plt.show()
```

```python
range_1 = range(4)    
range_2 = range(4)    

plt.scatter(range_1, range_2);
plt.show()
```

```python
range_1 = [3] * 4
range_2 = range(4)    

plt.scatter(range_1, range_2);
plt.show()
```

```python
range_1 = range(4) 
range_2 = [3] * 4    

plt.scatter(range_1, range_2);
plt.show()
```

# plt.title -> titlu

```python
x = [10, 10]
y = [8, 9]
plt.title("Acesta este titlul meu interesant")
plt.scatter(x, y);
```
