```python
import requests
```

```python
response = requests.get("https://dummyjson.com/recipes")
json_list = response.json()["recipes"]
```

```python
keys = list(json_list[0].keys())

filter_keys = [k for k in keys if type(json_list[0][k]) == int ]
filter_keys.remove("id")
filter_keys.remove("userId")
filter_keys.insert(0, "name")
```

```python
recipes = []
for item in json_list:
    rec =  { k:item[k] for k in filter_keys }
    recipes.append(rec)
    
```

```python
import pandas as pd

df = pd.DataFrame(recipes)
df.to_csv("recipes.csv")
```
